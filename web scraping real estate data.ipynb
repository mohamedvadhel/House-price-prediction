```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

url = "https://menzil.mr/"  

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

prices = []
divs1 = soup.find_all('div', class_='bottom-text-block')
for div in divs1:
    price = div.find('span', class_='price').text.strip()
    price = re.sub(r'[^\d]', '', price)  # Remove non-digit characters
    prices.append(price)

data1 = {'Price (MRO)': prices}
df = pd.DataFrame(data1)

data2 = {'Area (m²)': [], 'Location': []}
divs2 = soup.find_all('div', class_='container-card')
for div in divs2:
    area = div.find('div', class_='prop-area').text.strip()
    area = re.findall(r'\d+', area)[0]  # Extract the first number in the string
    location = div.find('div', class_='marker-text location').text.strip()
    data2['Area (m²)'].append(area)
    data2['Location'].append(location)

df2 = pd.DataFrame(data2)

df_combined = pd.concat([df, df2], axis=1)
df_combined.head(20)

# Save the dataframe as a CSV file
df_combined.to_csv('data.csv', index=False)
```

```python
import pandas as pd


# Create a dictionary to store the mapping of locations to zone numbers
zone_dict = {}

# Assign a unique zone number to each different location
zone_counter = 1
for location in df_combined['Location'].unique():
    zone_dict[location] = zone_counter
    zone_counter += 1

# Create a new column 'Zone Number' in the dataframe and assign the zone numbers
df_combined['Zone Number'] = df_combined['Location'].map(zone_dict)

print(df_combined.head(10))

# Print the zone dictionary
print(zone_dict)
```
